```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
df = pd.read_csv("/content/college_event_feedback.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Event_Name'].value_counts().plot(kind='bar')
plt.title("Feedback per Event")
plt.xlabel("Event_Name")
plt.ylabel("Number of Feedback")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Department'].value_counts().plot(kind='bar')
plt.title("Department Participant")
plt.xlabel("Department")
plt.ylabel("count")
plt.show()
```

```python
# Ensure feedback is treated as text
if 'Feedback' in df.columns:
    df['Feedback'] = df['Feedback'].astype(str)
```

```python
summary = {
    "Total Records": df.shape[0],
    "Total Columns": df.shape[1],
    "Missing Values": df.isnull().sum().sum(),
    "Duplicate Removed": "Yes"
}

pd.DataFrame(summary.items(), columns=["Metric", "Value"])
```

```python
df['Student_Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Event_Name', y='Rating', data=df)
plt.xticks(rotation=45)
plt.title("Event-wise Rating Distribution")
plt.show()
```

```python
import random
df['Rating'] = [random.randint(1,5) for _ in range(len(df))]

df.head()
```

```python
df['Feedback_Percentage'] = (df['Rating'] / 5) * 100
sns.histplot(df['Feedback_Percentage'], bins=10)
plt.title('Feedback Percentage Distribution')
plt.xlabel('Feedback Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Rating'].value_counts().sort_index().plot(kind='bar')
plt.title("Rating Distribution")
plt.xlabel("Rating (15)")
plt.ylabel("Number of Students")
plt.show()
```

```python
df.groupby('Event_Name')['Rating'].mean().plot(kind='bar')
plt.title("Average Rating per Event")
plt.xlabel("Event_Name")
plt.ylabel("Average Rating")
plt.ylim(0,5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Event_Name', data=df)
plt.title("Overall Event Popularity")
plt.xticks(rotation=45)
plt.show()
```
